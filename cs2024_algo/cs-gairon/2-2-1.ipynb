```python
def func1(n):
  s = 0
  for i in range(n):
    s+=1
  return s
```

```python
%%time
func1(1000000)
```

```python
def func2(n):
  s = 0
  for i in range(n):
    for j in range(n):
      s += i + j
```

```python
%%time
func2(10000)
```

```python
def func3(n):
  s = 0
  for i in range(1 << n):
    s+=i
```

```python
%%time
func3(30)
```
