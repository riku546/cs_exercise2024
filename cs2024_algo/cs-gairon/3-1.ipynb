```python
import random

def insert_sort(lst):
  for i in range(len(lst)):
    temp = lst[i]
    for j in range(i+ 1 , len(lst)):
      if temp > lst[j]:
        temp , lst[j]= lst[j] , temp
    lst[i] = temp
  return lst
  


```

```python
%%time
nums = [random.randint(0,100) for i in range(1000)]
print(insert_sort(nums))
```
