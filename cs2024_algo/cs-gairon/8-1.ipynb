```python

import matplotlib.pyplot as plt


principal = 100000 
rate = 0.02        
years = [i for i in range(1 , 51)]


simple_interest = [principal for i in years]
compound_interest = [principal + (principal * rate * i)  for i in years]



plt.figure(figsize=(10, 6))
plt.plot(years, simple_interest, label='Simple', linewidth=2)
plt.plot(years, compound_interest, label='Compound', linewidth=2)


plt.title('Simple and Compound interest')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)

plt.show()
```
