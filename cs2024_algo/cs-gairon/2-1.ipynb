```python
import math 
import matplotlib.pyplot as plt


x = [i for i in range(1 , 10001)]
nums= [math.log2(i) for i in x]
plt.plot(x , x ,label="Liner search")
plt.plot(x , nums , label="Binary search")
plt.title("Tokyo monthly temperature in 2020")
plt.ylabel("Time(ms)")
plt.xlabel("Length")
plt.legend()
plt.show()
```

```python


def gcd(a, b):
  remainder = a% b
  
  if(remainder == 0):
    print(b)
  else:
    gcd (b , remainder)
  
gcd(8177 , 3315)
```
